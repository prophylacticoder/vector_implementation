```python
#include <stdio.h>
#include <stdlib.h>

struct Vector {
    int* head;
    unsigned short int emptiness;
    unsigned int size;
    unsigned int cap;
};

struct Vector* init_vector(int *arr, unsigned int size) {
    struct Vector *vector = (struct Vector*) malloc(sizeof(struct Vector));
    unsigned int copied_elements = 0;
    // Allocates capacity of size * 2 
    vector->head = (int *) malloc(size * sizeof(int) * 2);
    if (vector->head == NULL) {
        return vector;
    }
    // Copies arr to new allocated vector
     for (; copied_elements < size; copied_elements++) {
         vector->head[copied_elements] = arr[copied_elements];
     }
    
     vector->size = copied_elements;
     vector->cap = size * 2;
     vector->emptiness = 0;
    
    return vector;
    
}

unsigned int vector_size(struct Vector* v) {
    return v->size;
}

unsigned int vector_cap(struct Vector* v) {
    return v->cap;
}

unsigned int vector_is_empty(struct Vector* v) {
    if (v->emptiness) {
        return 1;
    } 
    return 0;
}

int* vector_at(struct Vector* v, unsigned int at) {
    int *value = (int *) malloc(sizeof(int));
    if (v->head == NULL || at > v->size - 1) {
        return NULL;
    }
    *value = v->head[at];
    return value;
}

int main() {
    int intArray[10] = {1,2,3,4,5,6,7,8,9,0};
    unsigned int size = 10;
    struct Vector* intVector = init_vector(intArray, size);
    return 0;
}
```
